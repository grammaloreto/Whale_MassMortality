```python
#Dependencies
import pandas as pd
from fbprophet import Prophet
import matplotlib.pyplot as plt
```

**Monthly Chlorophyll concentration 2000-2022 in lon, lat = -75.5 , -45. Golfo de Penas. Chilean Patagonia**

```python
df = pd.read_csv('data/prophet_chl_monthly_2000-2022_whales.csv')
```

```python
df
```

```python
# Enabling yearly seasonality
# Train model
m = Prophet(interval_width=0.95, yearly_seasonality=True) # confidence interval of 95%
model = m.fit(df) # fit method
```

```python
# Forecast
future = m.make_future_dataframe(periods=100,freq='D')
forecast = m.predict(future)
forecast.head()
```

```python
forecast.tail(50)
```

```python
plot1 = m.plot(forecast)
plt.savefig('images/TimeSeries-chl_yearly-2000-2022.png', dpi=100)
```

```python
plt2 = m.plot_components(forecast)
plt.savefig('images/yearly-components_2000-2022.png', dpi=100)
```
